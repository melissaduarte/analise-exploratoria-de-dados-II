##SINESP

```python
import pandas as pd
import plotly.express as px
```

```python
sinesp = pd.read_excel('sinesp.xlsx')
```

```python
#descobrindo o tipo dos dados
sinesp.dtypes

```

```python
#olhando a base
sinesp.head(10)
```

```python
#excluir 2023

ano_a_remover = sinesp[(sinesp['Ano'] == 2023)].index

dados_sinesp = sinesp.drop(ano_a_remover)
dados_sinesp
```

```python
#excluir as UFs SP, MG, RJ, DF
estado_a_remover_sinesp = dados_sinesp[(dados_sinesp['UF'] == 'SP')|(dados_sinesp['UF'] == 'RJ')|(dados_sinesp['UF'] == 'MG')|(dados_sinesp['UF'] == 'DF')].index

sinesp_oficial = dados_sinesp.drop(estado_a_remover_sinesp)
sinesp_oficial
```

```python
#renomeando colunas

sinesp_oficial.rename(columns={'Raça/Cor': 'raça_cor'}, inplace=True)
sinesp_oficial.rename(columns={'Violência Sexual (SINESP)': 'Sinesp'}, inplace=True)
sinesp_oficial.rename(columns={'Faixa etária': 'faixa_etaria'}, inplace=True)
sinesp_oficial
```

```python
#groupby raca
sinesp_raca = sinesp_oficial.groupby(["Ano", "raça_cor"])[["Sinesp"]].sum().reset_index()
sinesp_raca
```

```python
import plotly.express as px
# Convertendo a coluna 'Ano' para string
sinesp_raca['Ano'] = sinesp_raca['Ano'].astype(str)
```

```python
#gráfico sinesp raça - linear
grafico_sinesp_raca_linear = px.line(sinesp_raca, x="Ano", y="Sinesp", color='raça_cor', markers=True,color_discrete_sequence=['#800080','#FFA500','#FFFF00','#008000','#0000FF','#FF0000'],
                       title="Número de Casos por Raça/Cor ao Longo dos Anos (Escala Linear) do Sinesp",
                       labels={"Sinesp": "Número de Casos (Sinesp)", "Ano": "Ano", "raça_cor": "Raça/Cor"})
grafico_sinesp_raca_linear.update_traces(text=sinesp_raca['Sinesp'], hoverinfo='text+x+y')
grafico_sinesp_raca_linear.update_yaxes(type='linear', tickmode='linear', tick0=0, dtick=500)  # Definindo intervalos de 500

# Exibindo o gráfico
grafico_sinesp_raca_linear.show()
```

```python
# Seu código para criar o gráfico
grafico_sinesp_raca_linear = px.line(
    sinesp_raca,
    x="Ano",
    y="Sinesp",
    color='raça_cor',
    markers=True,
    color_discrete_sequence=['#800080', '#FFA500', '#FFFF00', '#008000', '#0000FF', '#FF0000'],
    title="Número de Casos por Raça/Cor ao Longo dos Anos (Escala Linear) do Sinesp",
    labels={"Sinesp": "Número de Casos (Sinesp)", "Ano": "Ano", "raça_cor": "Raça/Cor"}
)

grafico_sinesp_raca_linear.update_traces
```

```python
#groupby 2: escolaridade
sinesp_escolaridade = sinesp_oficial.groupby(["Ano", "Escolaridade"])[["Sinesp"]].sum().reset_index()
sinesp_escolaridade
```

```python
#gráfico sinesp escolaridade - linear

sinesp_escolaridade['Ano'] = sinesp_escolaridade['Ano'].astype(str)

grafico_sinesp_escolaridade = px.line(sinesp_escolaridade, x="Ano", y="Sinesp", color='Escolaridade', markers = True,color_discrete_sequence=['#FF7F7F','#779ECB','#77DD77','#CBAACB','#FFB347'],
              title="Número de Casos por Escolaridade ao Longo dos Anos (Escala Linear) do SINESP",
              labels={"Sinesp": "Número de Casos (Sinesp)", "Ano": "Ano", "escolaridade": "Escolaridade"})
grafico_sinesp_escolaridade.update_traces(text=sinesp_escolaridade['Sinesp'], hoverinfo='text+x+y')
grafico_sinesp_escolaridade.update_yaxes(type='linear')
```

```python
#groupby 3: faixa etária
sinesp_faixa = sinesp_oficial.groupby(["Ano", "faixa_etaria"])[["Sinesp"]].sum().reset_index()
sinesp_faixa
```

```python
#gráfico faixa etária sinesp - escala linear

sinesp_faixa['Ano'] = sinesp_faixa['Ano'].astype(str)

grafico_sinesp_faixa = px.line(sinesp_faixa, x="Ano", y="Sinesp", color='faixa_etaria', markers = True,
              title="Número de Casos por Faixa Etária ao Longo dos Anos (Escala Linear) do SINESP",
              labels={"Sinesp": "Número de Casos (Sinesp)", "Ano": "Ano", "faixa_etaria": "Faixa Etária"})
grafico_sinesp_faixa.update_traces(text=sinesp_faixa['Sinesp'], hoverinfo='text+x+y')
grafico_sinesp_faixa.update_yaxes(type='linear')
```

```python
sinan = pd.read_excel('sinan.xlsx')
```

```python
#descobrindo o tipo dos dados
sinan.dtypes
```

```python
sinan.head(10)
```

```python
ano_sinan_remover = sinan[( sinan['Ano'] == 2011) | ( sinan['Ano'] == 2012) | ( sinan['Ano'] == 2013)  | ( sinan['Ano'] == 2014) | ( sinan['Ano'] == 2015) | ( sinan['Ano'] == 2016) |( sinan['Ano'] == 2017) | ( sinan['Ano'] == 2023)].index

dados_sinan = sinan.drop(ano_sinan_remover)
dados_sinan
```

```python
estado_a_remover = dados_sinan[(dados_sinan['UF'] == 'SP')|(dados_sinan['UF'] == 'RJ')|(dados_sinan['UF'] == 'MG')|(dados_sinan['UF'] == 'DF')].index

sinan_oficial = dados_sinan.drop(estado_a_remover)
sinan_oficial
```

```python
sinan_oficial.rename(columns={'Raça/Cor': 'raça_cor'}, inplace=True)
sinan_oficial.rename(columns={'Violência Sexual (SINAN)': 'Sinan'}, inplace=True)
sinan_oficial.rename(columns={'Faixa etária': 'faixa_etaria'}, inplace=True)
sinan_oficial
```

```python
#groupby 1: raça
sinan_raca = sinan_oficial.groupby(["Ano", "raça_cor"])[["Sinan"]].sum().reset_index()
sinan_raca
```

```python
#gráfico sinan raça - linear
sinan_raca['Ano'] = sinan_raca['Ano'].astype(str)
grafico_sinan_raca_linear = px.line(sinan_raca, x="Ano", y="Sinan", color='raça_cor', markers=True, color_discrete_sequence=['#800080','#FFA500','#FFFF00','#008000','#0000FF','#FF0000'],
              title="Número de Casos por Raça/Cor ao Longo dos Anos (Escala Linear) do SINAN",
              labels={"Sinan": "Número de Casos (Sinan)", "Ano": "Ano", "raça_cor": "Raça/Cor"})
grafico_sinan_raca_linear.update_traces(text=sinan_raca['Sinan'], hoverinfo='text+x+y')
grafico_sinan_raca_linear.update_yaxes(type='linear')

# Exibindo o gráfico
grafico_sinan_raca_linear.show()


```

```python
#groupby 2: escolaridade
sinan_escolaridade = sinan_oficial.groupby(["Ano", "Escolaridade"])[["Sinan"]].sum().reset_index()
sinan_escolaridade
```
